```python
# create a list of our conditions
conditions = [
    ((df1['Type (Hidden)'] == 'Call') | (df1['Call Duration In Seconds (seconds)'] >= 1)),
    ((df1['Type (Hidden)'] == 'Email') & (df1['Call Duration In Seconds (seconds)'] == 0)),
    (df1['Type (Hidden)'] == 'Action')
    ]

# create a list of the values we want to assign for each condition
values = ['Call', 'Email','Action']

# create a new column and use np.select to assign values to it using our lists as arguments
df1['Type (Hidden)'] = np.select(conditions, values)
```

## We are creating a new dataframe.
## Creating a new column "NEW_DATE_COLUMN". We will pivot the entire dataset using said column. This is done as currently an individual has mulltiple date time columns. We want to only have one in each row. The reason for this is that we cant assign a funnel stage to a customer if they have experienced more then one funnel stage. We can only assign either the first or last funnel stage. Therefore, by pivoting the data on the new column we can assign a date time for each funnel stage a customer has completed.

## NEW_DATE_COLUMN was created by combining all the 5 unique date columns wihin the dataframe as a list.

```python
df1['NEW_DATE_COLUMN'] = df1[["Date New Enquiry","Date Prospect","Date App Started","Date App Submitted","Date Registered / Deposit Paid"]].values.tolist()
```

```python
df1['NEW_DATE_COLUMN'] = df1['NEW_DATE_COLUMN'].values.tolist()
```

## Seperate the NEW_DATE_COLUMN column list to get rows with only 1 date for each funnel type.

```python
df1 = df1.explode('NEW_DATE_COLUMN')
```

```python
display(df1.drop_duplicates())
```

## Remove the previous 5 Dates as they are no longer needed.

```python
df1 = df1.drop(['Date New Enquiry'], axis=1)
df1 = df1.drop(['Date Prospect'], axis=1)
df1 = df1.drop(['Date App Started'], axis=1)
df1 = df1.drop(['Date App Submitted'], axis=1)
df1 = df1.drop(['Date Registered / Deposit Paid'], axis=1)
```

## Drop duplicated values within the new dataframe. This is done as some individuals will have duplicated rows. For instance if Customer only completed the Enquiry stage they would have 4 extra rows with date assigned as 0 as they have not completed any other funnel stage. Therefore, we simply remove them.

```python
a = df1.drop_duplicates()
```

```python
a = a[a['NEW_DATE_COLUMN'] > 0]
```

```python
a
```

## Creating a new column called Streak. This will be used to assign the funnel type to each unique row.Therefore, Streak simply shows us the consecutive occurance for an individual. So if Customer 1 completed the Registered / Deposit Paid stage we will asign a Streak of 1,2,3,4,5 for each consecutive row. Therefore, row 1 will have 1, row 2 will have 2 and so on. 

```python
a = a.sort_values(by=['Activity ID', 'NEW_DATE_COLUMN']).reset_index()
a['Streak'] = a['Activity ID'].groupby((a['Activity ID'] != a['Activity ID'].shift()).cumsum()).cumcount() + 1
a
```

## Now we will use the newly created Streak collumn and assign a funnel stage for each row. Therefore, if row contains Streak = 1 we assign Enquiry, if 2 we assign Prospect and so on.

```python
# create a list of our conditions
conditions = [
    (a['Streak'] == 1),
    (a['Streak'] == 2),
    (a['Streak'] == 3),
    (a['Streak'] == 4),
    (a['Streak'] == 5)
    ]

# create a list of the values we want to assign for each condition
values = ['Enquiry', 'Prospect','App Started','App Submitted','Registered / Deposit Paid']

# create a new column and use np.select to assign values to it using our lists as arguments
a['Funnel_Stage'] = np.select(conditions, values)
```

## Lastly we drop all of the column which we do not need. We will drop all of the previous funnel stage columns, as weel as the newsly created Streak column.

```python
a = a.drop(['index'], axis=1)
a = a.drop(['Enquiry'], axis=1)
a = a.drop(['Prospect'], axis=1)
a = a.drop(['App Started'], axis=1)
a = a.drop(['App Submitted'], axis=1)
a = a.drop(['Registered / Deposit Paid'], axis=1)
a = a.drop(['Streak'], axis=1)
```
